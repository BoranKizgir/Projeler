```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("imdb_top_1000.csv")

df["Runtime"] = df["Runtime"].str.replace(" min", "").astype(int)

df["Released_Year"] = pd.to_numeric(df["Released_Year"], errors="coerce")
film_sayisi_yillara_gore = df["Released_Year"].value_counts().sort_index()

plt.figure(figsize=(12, 4))
film_sayisi_yillara_gore.plot(kind="bar", color="lightblue")
plt.title("Yllara Gre Film Says")
plt.xlabel("Yl")
plt.ylabel("Film Says")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
df["IMDB_Rating"].plot(kind="hist", bins=20, color="orange", edgecolor="black")
plt.title("IMDb Puan Dalm")
plt.xlabel("IMDb Puan")
plt.ylabel("Film Says")
plt.tight_layout()
plt.show()

df["Ana_Tur"] = df["Genre"].str.split(",").str[0]
populer_turler = df["Ana_Tur"].value_counts().head(10)

plt.figure(figsize=(8, 4))
populer_turler.plot(kind="bar", color="green")
plt.title("En Popler 10 Film Tr")
plt.xlabel("Tr")
plt.ylabel("Film Says")
plt.tight_layout()
plt.show()

en_iyi_filmler = df.sort_values("IMDB_Rating", ascending=False).head(10)
print("En Yksek IMDb Puanna Sahip 10 Film:")
print(en_iyi_filmler[["Series_Title", "IMDB_Rating"]])
```
